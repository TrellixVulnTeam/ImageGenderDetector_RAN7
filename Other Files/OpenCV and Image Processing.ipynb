For an image, 
pixels are the values from range 0 -> (2^n - 1)

eg: for an 8 bit image: pixels are: 0 -> (2^8 - 1) = 0 -> 255

#**Working with Videos:**

**Video:** Different from images, they are combination of images, like 24 frames per second, i.e. in a second, 24 images are coming.

*Note:* It is difficult to see the image, so we use a while loop to loop through the frames and display them.

Steps to work with video.
1. Load a video using cv2.VideoCapture("path")
2. Use same functionalities used in the image above.

```python
def detect_faces(img):
  # convert to gray scale
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  #cv2_imshow(gray)

  # use haar cascade classfier
  haar = cv2.CascadeClassifier("haarcascade_classifier.xml")

  faces = haar.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

  for x, y, w, h in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 255), 2)
  return img
```

```python
imgg = detect_faces(img)
cv2.imshow('face', imgg)
cv2.waitKey(0)
```

```python
# step 1: Load the video.
cap = cv2.VideoCapture("images/testV.mp4")

while True: 
  ret, frame = cap.read() 
  if ret == False: 
    break 

  frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
  frame = detect_faces(frame)
  
  
  cv2.imshow('Object', frame)

  # we need to control our video for 'n' frames per second, thus for every 40 ms run the loop.
  if cv2.waitKey(1) == 27: 
    break 
cv2.destroyAllWindows() 
cap.release()
```
