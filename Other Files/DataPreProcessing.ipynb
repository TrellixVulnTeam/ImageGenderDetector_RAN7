**Now Apply this to all the images:**

Since there are around 14k images, it is not advisable to write loops.
Instead we will write a function to do so.

```python
# Apply to all 14000 images.
def extract_images(path, gender, i):
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = haar.detectMultiScale(gray, 1.5, 5)
    for x, y, w, h in faces:
        roi = img[y:y+h, x:x+w]
        cv2.imwrite('images/cropped_images/{}/{}.png'.format(gender, i), roi)
```

```python
extract_images(femalepath[0], 'female', 1)
```

```python
for i, path in enumerate(femalepath):
    try:
        extract_images(path, 'female', i)
        print('INFO: {}/{} processed successfully'.format(i, len(femalepath)))
    except:
        #This is used since some images may not be processed successfully.
        print('INFO: {}/{} not processed'.format(i, len(femalepath)))
```

```python
for i, path in enumerate(malepath):
    try:
        extract_images(path, 'male', i)
        print('INFO: {}/{} processed successfully'.format(i, len(malepath)))
    except:
        #This is used since some images may not be processed successfully.
        print('INFO: {}/{} not processed'.format(i, len(malepath)))
```

# Resizing images:
**Steps:**
1. Read image
2. Convert the image to grayscale
3. Resize into a 100 x 100 array
4. Flatten the image. -> Convert into 1x10000 sized array

```python
def resize(pathToResize):
    try:
        # Step1
        img = cv2.imread(pathToResize)

        # Step2
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Step3
        size = gray.shape[0]
        if size >= 100:
            gray_re = cv2.resize(gray, (100, 100), cv2.INTER_AREA)   #Shrink
        else:
            gray_re = cv2.resize(gray, (100, 100), cv2.INTER_CUBIC)   #Enlarge

        flat_img = gray_re.flatten()
        return flat_img
    except:
        return None
```

# Data Preprocessing
1. How to load the data and look into datatypes.
2. Detect and Remove missing values.
3. Data Normalization (min and max scaling (Since the data does not contains many outliers))
Note: Data Normalization is used to convert and normalize all the data in a specific range. Eg: 0 -> 1

```python
# Load the pickle file we saved earlier.
df = pickle.load(open('images/data_frame_images_100_100.pickle', 'rb'))
#rb stands for read bytes.
```

```python
df.head()
```

```python
df.info()
```

```python
# Check for missing values:
df.isnull().sum()
```

```python
# Removing missing values. (axis=0 means rows)
df.dropna(axis=0, inplace=True)
```

```python
# Data normalization: Applies only to input values.

# Split the data into two parts.
X = df.iloc[:,1:].values        # Independent features.
y = df.iloc[:,0].values         # Dependent features

X.shape, y.shape
```

**Min max scaling for data normalization:**

Xnorm = $\frac {x - minValue}{maxValue - minValue}$

Since our data is all 8 bit images here, it is expected that the values are: 0 -> 2^((n=8)-1)

```python
X.min(), X.max()
# Range for an 8 bit image is 0-> 255
```

```python
Xnorm = X/X.max()
Xnorm

# We can see all our data is between 0 - 1
```
